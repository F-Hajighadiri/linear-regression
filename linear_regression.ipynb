```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([[740, 724, 739], [724, 739, 727], [739, 727, 743], [727, 743, 736], [743, 736, 725],
              [736, 725, 731], [725, 731, 743], [731, 743, 740], [743, 740, 754], [740, 754, 775]])
y = np.array([727, 743, 736, 725, 731, 743, 740, 754, 775, 760]).reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(x, y)

y_predict = model.predict(x)
```

```python
plt.plot(x[:, 0], y, 'o', label='True feature 1')
plt.plot(x[:, 0], y_predict, 'o', label='predicted feature 1')
plt.legend()
```

```python
plt.plot(x[:, 1], y, 'o', label='True feature 2')
plt.plot(x[:, 1], y_predict, 'o', label='predicted feature 2')
plt.legend()
```

```python
plt.plot(x[:, 2], y, 'o', label='True feature 3')
plt.plot(x[:, 2], y_predict, 'o', label='predicted feature 3')
plt.legend()
```

```python
plt.plot(y, y_predict, '*')
```

```python
x_smp = np.array([[740, 724, 739], [110, 100, 200], [600, 670, 300], [590, 540, 400]])  
y_smp = model.predict(x_smp)
print(y_smp)
```
